```python
import streamlit as st
import pandas as pd
import joblib
import os

# Caminho seguro do modelo
BASE_DIR = os.path.dirname(os.path.abspath(__file__))
model_path = os.path.join(BASE_DIR, "modelo.pkl")

model = joblib.load(model_path)

st.title("Predição de Nível de Obesidade")
st.markdown("Preencha as informações abaixo:")

Gender = st.selectbox("Genero", ["Male", "Female"])
Age = st.number_input("Idade", min_value=1, max_value=100, value=25)
Height = st.number_input("Altura (metros)", min_value=1.0, max_value=2.5, value=1.70)
Weight = st.number_input("Peso (kg)", min_value=30.0, max_value=200.0, value=70.0)
family_history = st.selectbox("Histórico Familiar de Obesidade", ["yes", "no"])
FAVC = st.selectbox("Consumo de alimentos calóricos frequentemente (FAVC)", ["yes", "no"])
FCVC = st.number_input("Consumo de vegetais frequentemente (FCVC)", min_value=0.0, max_value=3.0, value=2.0)
NCP = st.number_input("Numero de Refeições (NCP)", min_value=1.0, max_value=5.0, value=3.0)
CAEC = st.selectbox("Lanches entre refeições (CAEC)", ["no", "Sometimes", "Frequently", "Always"])
SMOKE = st.selectbox("Voce Fuma?", ["yes", "no"])
CH2O = st.number_input("Consumo de agua em Litros (CH2O)", min_value=0.0, max_value=3.0, value=2.0)
SCC = st.selectbox("Monitora o consumo de calorias (SCC)", ["yes", "no"])
FAF = st.number_input("Frequencia de atividade física (FAF)", min_value=0.0, max_value=3.0, value=1.0)
TUE = st.number_input("Tempo de tela durante o dia (TUE)", min_value=0.0, max_value=3.0, value=1.0)
CALC = st.selectbox("Consome alcool (CALC)", ["no", "Sometimes", "Frequently", "Always"])
MTRANS = st.selectbox("Tipo de transporte (MTRANS)", ["Walking", "Bike", "Motorbike", "Public_Transportation", "Automobile"])

if st.button("Prever"):

    dados = pd.DataFrame([{
        "Gender": Gender,
        "Age": Age,
        "Height": Height,
        "Weight": Weight,
        "family_history": family_history,
        "FAVC": FAVC,
        "FCVC": FCVC,
        "NCP": NCP,
        "CAEC": CAEC,
        "SMOKE": SMOKE,
        "CH2O": CH2O,
        "SCC": SCC,
        "FAF": FAF,
        "TUE": TUE,
        "CALC": CALC,
        "MTRANS": MTRANS
    }])

    pred = model.predict(dados)

    st.success(f"Nível previsto: {pred[0]}")
```
